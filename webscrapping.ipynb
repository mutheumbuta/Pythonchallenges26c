```python

import requests
from bs4 import BeautifulSoup
import pandas as pd

#  Fetch BBC News homepage
url = "https://www.bbc.com/news"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extract headlines and summaries
articles = []
for item in soup.find_all("div", class_="gs-c-promo", limit=15):
    headline = item.find("h3")
    summary = item.find("p")
    if headline:
        articles.append({
            "Headline": headline.get_text(strip=True),
            "Summary": summary.get_text(strip=True) if summary else ""
        })

#Export to CSV
df = pd.DataFrame(articles)
df.to_csv("bbc_headlines.csv", index=False)

print("Export complete! File saved as bbc_headlines.csv")
```
